```python
#Reemplace los String â--" por valores nulos
valores_faltants = ["--"]
df = pd.read_csv('/content/drive/MyDrive/Admindatos/data_corregida.csv', na_values = valores_faltants)
df.head()
```

```python
#Reemplazar los valores perdidos inesperados de la columna âOverTimeâ por No
#Convertir valores inesperados a valores nulos
cnt=0
for row in df['OverTime']:
    try:
        int(row)
        df.loc[cnt, 'OverTime']=np.nan
    except ValueError:
        pass
    cnt+=1
```

```python
#Cambio de Nulo a No en la columna OverTime
df = df.fillna({'OverTime': 'No'})
```

```python
#Eliminar aquella columna (s) que tengan mÃ¡s del 70% de sus valores en nulo
#Ver columnas con mas de 70% de valores nulo
filtro = (df.isnull().sum() / len(df))*100 > 70
print(filtro)
```

```python
#Eliminar columna con valor mayor a 70%
df = df.loc[:, df.isnull().mean() < .7]
df.head()
```

```python
#Eliminar aquellas filas que tenga menos de dos elementos.
df = df.dropna(axis=0, thresh=3)
df
```

```python
#Reemplazar los valores nulos de la columna de âGenderâ por U (Unknown) para aquellos empleados 
#que no especificaron este campo.
df = df.fillna({'Gender': 'U'})
```

```python
#Reemplazar los valores nulos de la columna âDepartmentâ por âBenchâ
df = df.fillna({'Department': 'Bench'})
```

```python
#Reemplazar los valores nulos de la columna âHourlyRateâ por la mediana
mediana = df['HourlyRate'].median()
df['HourlyRate'].fillna(mediana, inplace=True)
```

```python
#Salvar el dataframe una vez realizado el procesamiento de los anteriores
df.to_csv('/content/drive/MyDrive/Admindatos/data_final.csv',header=False)
```
